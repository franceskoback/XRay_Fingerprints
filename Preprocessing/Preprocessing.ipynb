```python
#names = set(df.loc[:,"LOC"])
def text_file_gen(df,filter_var):
    names=df[filter_var].unique()
    print(names)
    for name in names:
        variable = df[df[filter_var]==name]
        variable=list(variable.index)
        with io.open(str(name) + ".txt", 'w') as f:
            for item in variable:
                f.write("%s\n" % item)
            print("Done with " + str(name))
        print(len(variable))
#text_file_gen(df,filter_var)
```
